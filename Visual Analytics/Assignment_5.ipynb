```python
import requests
from bs4 import BeautifulSoup

# Replace this URL with your desired URL
url = "https://www.newsweek.com/millennials-gen-z-workplace-pronouns-emails-1797270"

# Set a custom User-Agent header to mimic a browser
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.3'
}

# Send a GET request to the URL with the custom headers
response = requests.get(url, headers=headers)

# Check if the request was successful (status code 200)
if response.status_code == 200:
    # Parse the HTML content of the page
    soup = BeautifulSoup(response.content, "html.parser")

    # Example: Extract numbers from paragraphs (change this based on the actual HTML structure)
    paragraphs = soup.find_all('p')
    numbers = []
    for paragraph in paragraphs:
        # Check if the paragraph contains a number (you might need to adjust this condition)
        if any(char.isdigit() for char in paragraph.text):
            numbers.append(paragraph.text.strip())

    # Print the extracted numbers
    print("Numbers:")
    for number in numbers:
        print(number)
else:
    print(f"Error: Unable to fetch the page. Status code: {response.status_code}")
```

```python
import matplotlib.pyplot as plt

# Sample data (replace with actual data from your description)
age_groups = ['25-44 Millennials', '18-24 Gen Z', '45+ Age 45 or more', '18-29 Young Americans', '30-49 Age 30-49', '50+ Age 50 or above']
support_percentages = [60, 40, 30, 33, 19, 0]
oppose_percentages = [30, 50, 50, 77, 81, 0]
know_someone_percentages = [65, 40, 0, 33, 19, 0]

# Plotting the bar chart for support and opposition percentages
bar_width = 0.35
index = range(len(age_groups))

fig, ax = plt.subplots()
bar1 = ax.bar(index, support_percentages, bar_width, label='Support')
bar2 = ax.bar([i + bar_width for i in index], oppose_percentages, bar_width, label='Oppose')

# Adding labels and title
ax.set_xlabel('Age Groups')
ax.set_ylabel('Percentage')
ax.set_title('Support and Opposition to Gender Pronouns in Work Emails by Age Group')
ax.set_xticks([i + bar_width/2 for i in index])
ax.set_xticklabels(age_groups, rotation=45, ha='right')
ax.legend()

# Display the bar chart
plt.tight_layout()
plt.show()

# Plotting the line chart for knowing someone using gender-neutral pronouns --- I did not consider all the columns just [3:5] because I have real data on these 3 groups
fig, ax = plt.subplots()
ax.plot(age_groups[3:5], know_someone_percentages[3:5], marker='o', label='Know Someone')

# Adding labels and title
ax.set_xlabel('Age Groups')
ax.set_ylabel('Percentage')
ax.set_title('Percentage of People Knowing Someone Using Gender-Neutral Pronouns')
ax.legend()

# Display the line chart
plt.tight_layout()
plt.show()
```
