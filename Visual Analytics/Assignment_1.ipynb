```python
import pycountry
from sklearn.preprocessing import MinMaxScaler

gdp_columns = df.columns[15:-1]  # Assuming the GDP columns start at index 15 and -1 not last column 'Unnamed'

country_names = [country.name for country in pycountry.countries]

# Filter the DataFrame based on the list of country names, making a copy explicitly
country_df = df[df['Country Name'].isin(country_names)].copy()

# Calculate the average GDP for each country
country_df['Average GDP'] = country_df[gdp_columns].mean(axis=1)

# Normalize the 'Average GDP' values to be in the range [0, 25000]
scaler = MinMaxScaler(feature_range=(0, 25000))
scaled_gdp = scaler.fit_transform(country_df['Average GDP'].values.reshape(-1, 1))

# Convert scaled values to integers
country_df['Average GDP'] = scaled_gdp.astype(int)

# Select relevant columns for display
result_df = country_df[['Country Name', 'Average GDP']]

# Sort DataFrame by average GDP in descending order
result_df_sorted = result_df.sort_values(by='Average GDP', ascending=False)

# Select the top 5 rows
top_5_countries = result_df_sorted.head()

top_5_countries
```

```python
# Plotting
import matplotlib.pyplot as plt
from matplotlib.ticker import FuncFormatter
plt.figure(figsize=(12, 8))

# Iterate through each country and plot its GDP over the years for 5 richest country
for index, row in top_5_countries.iterrows():
    country_name = row['Country Name']
    country_data = country_df.loc[country_df['Country Name'] == country_name, gdp_columns].values.flatten()

    # Plot the line
    line, = plt.plot(gdp_columns, country_data, label=country_name)

    # Add a filled circle at the last data point with the same color as the line
    plt.scatter(gdp_columns[-1], country_data[-1], color=line.get_color(), marker='o', s=100)

# Show the legend
plt.legend()

# Filter other countries excluding the target country
other_countries_df = country_df[~country_df['Country Name'].isin(top_5_countries['Country Name'])]

# Filter the last year data based on the condition
filtered_last_year_data = other_countries_df[gdp_columns[-2]].between(0, 25000)

# Plot the last year data of other countries with gray color
for index, row in other_countries_df[filtered_last_year_data].iterrows():
    country_name = row['Country Name']
    country_data = row[gdp_columns].values.flatten()

    plt.plot(gdp_columns, country_data, color='#CCCCCC', label=country_name, alpha=0.7)

# Customize the x-axis and y-axis
selected_years = ['1980', '1990', '2000', '2010', '2020']
plt.xticks(selected_years, rotation=45)

formatter = FuncFormatter(lambda x, _: f'{x / 1e9:.0f}')
plt.gca().yaxis.set_major_formatter(formatter)

# does not work too
#custom_y_ticks = [0, 5000, 10000, 15000, 20000, 25000]
#plt.yticks(custom_y_ticks)
# Set y-axis limits to ensure all custom ticks are displayed
#plt.ylim(bottom=0, top=25000)
# Customize the plot

title_line1 = r"$\bf Evolution of the 20 Richest Counties GDP over the Past of 50 Years$"
title_line2 = 'Focus on current 5 richest countries from 1973 to 2022'
plt.title(f'{title_line1}\n{title_line2}')


plt.xlabel('Year of Record', fontweight='bold', fontsize=14)  # font size
plt.ylabel('GDP (Billions USD)', fontsize=14)  # font size

plt.grid(True)

target_country = 'China'
target_country_data = df[df['Country Name'] == target_country][gdp_columns].values.flatten()

# Calculate the center coordinates
center_x = (plt.xlim()[0] + plt.xlim()[1]) / 2
center_y = (plt.ylim()[0] + plt.ylim()[1]) / 2

# Add a text annotation with a curved arrow

arrow_text = 'During in 2000s\nChina began experiencing rapid economic growth.\noutpacing all other countries.'
plt.annotate(arrow_text, xy=(gdp_columns[-1], target_country_data[-1]),
             xytext=(center_x, center_y),  # Point to the center coordinates
             arrowprops=dict(facecolor='black', arrowstyle='->', connectionstyle='arc3,rad=0.3', lw=1),
             fontsize=12, ha='center')


# Show the plot
plt.show()
```
